```python
import os

import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv("../../.env")

engine = create_engine(
    f"postgresql+psycopg://{os.environ.get('DATABASE_USERNAME')}:{os.environ.get('DATABASE_PASSWORD')}@localhost:5432/{os.environ.get('DATABASE_NAME')}")

schools_to_fetch = pd.read_sql(
    "SELECT e.* from education e LEFT JOIN education_acara ea on e.id = ea.education_id WHERE acara_id is null AND is_high_school",
    engine)
schools_to_fetch
```

```python
acara_schools = pd.read_sql(
    '''SELECT * from acara_school_locations_2022 WHERE "school type" in ('Secondary', 'Combined');''', engine)
acara_schools
```

```python
# This is fixed now
merged = pd.merge(schools_to_fetch, acara_schools, how="inner", left_on=schools_to_fetch.school_name,
                  right_on=acara_schools["school name"], suffixes=(None, "_acara"))
merged.drop_duplicates(subset="school_name", inplace=True, keep=False)
merged[["id", "acara sml id"]].rename({"id": "education_id", "acara sml id": "acara_id"}, axis=1).to_sql(
    "education_acara", engine, if_exists="append", index=False)
```

```python
import os

import geopandas as gpd
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv("../../.env")

engine = create_engine(
    f"postgresql+psycopg://{os.environ.get('DATABASE_USERNAME')}:{os.environ.get('DATABASE_PASSWORD')}@localhost:5432/{os.environ.get('DATABASE_NAME')}")

schools_to_fetch = gpd.read_postgis(
    "SELECT e.* from education e LEFT JOIN education_acara ea on e.id = ea.education_id WHERE acara_id is null AND is_high_school",
    engine, geom_col="geometry")
acara_schools = gpd.read_postgis('''SELECT * from acara_school_locations_2022 WHERE "school type" in ('Secondary', 'Combined');''', engine
)
acara_schools = acara_schools.to_crs("EPSG:9473")
schools_to_fetch = schools_to_fetch.to_crs("EPSG:9473")
matched_dist = schools_to_fetch.sjoin_nearest(acara_schools, how="left", max_distance=1000)
matched_dist[matched_dist["acara sml id"].notna()][["id", "acara sml id"]].rename({"id": "education_id", "acara sml id": "acara_id"}, axis=1).to_sql(
    "education_acara", engine, if_exists="append", index=False)
```
